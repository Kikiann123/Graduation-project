```python
"""Driver safety estimation"""

#load libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

##supress scientific notification

pd.options.display.float_format = '{:.4f}'.format

#read data from file
data = pd.read_csv("C:/Users/kikia/OneDrive/Master Smart Systems Engineering/Graduation/Afstudeerproject/Dataset google takeout/Dataset2.txt")
data.columns = ['DriverId', 'EventName', 'Latitude', 'Longitude', 'Speed', 'ts']
#data = data.loc[data['EventName']>0, ['DriverId', 'EventName', 'Latitude', 'Longitude', 'Speed', 'ts']].groupby(['DriverId', 'EventName', 'Latitude', 'Longitude', 'Speed', 'ts'], as_index=False).mean()
print(data.head(10))

#lets observe some rows
data.head(10)
print(data.head)

#Time in which samples were collected:
#maxTime = data['ts'].max()
#minTime = data['ts'].min()
#print('Events were collected between' + minTime + "and" + maxTime)

#number of data
dataCount = len(data.index.value_counts())
print("number of data: {}".format(dataCount))

#number of events per driver
eventsPerDriver = data.groupby('DriverId').agg({"EventName": "count"}).add_suffix('_count')
print("Mean number of events per driver {:0.1f}".format(np.mean(eventsPerDriver).values[0]))
eventsPerDriver.head(10)

#Events per subscriber
eventsPerDriver.sort_values(by='EventName_count', ascending=False, inplace=True)

eventsHist = eventsPerDriver['EventName_count'].hist()
eventsHist.set_title("Distribution of events per driver")
eventsHist.set_xlabel("Number of events")
eventsHist.set_ylabel("Number of drivers")

#Events type
dataReasonHist = data.groupby(['EventName'])[['EventName']].agg('count').add_suffix('_Count').reset_index().sort_values('EventName_Count')
dataReasonHist.plot.barh('EventName', 'EventName_Count', figsize=(18, 8), fontsize=10)
plt.show()

#removing distance and timed network events
events = [event for event in data.EventName if event not in ['Distance Event', 'Timed Event']]
newCount = len(data[data['EventName'].isin(events)])
print(events)

#Data prep
RELEVANT_EVENTS = ['Harsh acceleration', 'Harsh braking', 'Harsh turn', 'Reached max speed', 'Out of max speed']

def prepData(data, minRecordsPerSubscriber = 50):
    data.reset_index(inplace=True)
    print(" Starting data prep. Length:", len(data))

    #Remove Nas
    #data = data.dropna()
    #print("Removed NA.s Length:", len(data))

    #filter out unwanted events
    data1 = data[data.EventName.isin(RELEVANT_EVENTS)]
    print("Only events that are relevant for modeling. Length:", len(data1))

    #Filter out users with too few samples
    #eventCountPerDriver = data1.groupby('DriverId')['DriverId'].agg('count')
    #driversWithManyRecords = eventCountPerDriver[eventCountPerDriver > minRecordsPerSubscriber]
    #driversWithManyRecords.keys()
    #data1 = data1[data1.DriverId.isin(driversWithManyRecords.keys())]
    #print('Filtering uses with too few samples. Length:', len(data1))

    print("Task executed")
    return(data1)

data1 = prepData(data)
data1.head()
print(data1.head())

relevantEventsPerSubscriber = data1.groupby('DriverId').agg({"EventName": "count"}).sort_values(by='EventName', ascending=False)

eventsHist = relevantEventsPerSubscriber['EventName'].hist()
eventsHist.set_title("Distribution of events per driver")
eventsHist.set_xlabel("Number of events")
eventsHist.set_ylabel("Number of drivers")
plt.show()

#types of events
data1ReasonHist = data1.groupby('EventName')[['EventName']].agg('count').add_suffix('_Count').reset_index().sort_values('EventName_Count',ascending=True)
data1ReasonHist.plot.barh('EventName', 'EventName_Count',figsize=(10,4), fontsize=10)
plt.show()

#calculate total distance the driver had travelled
def haversine(lon1, lat1, lon2, lat2):
    lon1, lat1, lon2, lat2 = map(np.radians, [lon1, lat1, lon2, lat2])

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = np.sin(dlat/2.0)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2.0)**2
    c = 2 * np.arcsin(np.sqrt(a))
    km = 6367 * c
    return km

def total_distance(oneDriver):
    dist = haversine(oneDriver.Longitude.shift(1), oneDriver.Latitude.shift(1),
                     oneDriver.loc[1:,'Longitude'], oneDriver.loc[1:, 'Latitude'])
    return np.sum(dist)

def calculate_overall_distance_travelled(data):
    data1DistancePerDriver = data.groupby('DriverId').apply(total_distance).reset_index(name='Distance')
    return data1DistancePerDriver

distancePerDriver = calculate_overall_distance_travelled(data1)
print(distancePerDriver.head(10))

"""Feature engineering"""

#Define feature set
def create_feature_set(data1, distancePerDriver):
    data1EventAggByDriver = data1.groupby(['DriverId', 'EventName'])[['EventName']].agg('count').add_suffix('_Count').reset_index()

    #pivot events into columns
    data1EventMatrix = data1EventAggByDriver.pivot(index ='DriverId', columns='EventName', values='EventName_Count').add_prefix('F_').fillna(0).reset_index()
    data1EventMatrix = data1EventMatrix.merge(distancePerDriver, how='inner', on='DriverId')
    data1EventMatrix.set_index('DriverId', inplace=True)

    featureCols = [col for col in data1EventMatrix if col.startswith('F_')]
    data1EventMatrix[featureCols] = data1EventMatrix[featureCols].div(data1EventMatrix['Distance'], axis=0)

    data1FeatureSet = data1EventMatrix[featureCols]

    return data1FeatureSet

features = create_feature_set(data1, distancePerDriver)
print('features dataframe:', features.head(10))

features = features[['F_Harsh acceleration', 'F_Harsh braking', 'F_Harsh turn']]
features.head()
print('renamed features:', features.head())

#scatterplot matrix of features
#sns.set(style="ticks", font_scale=1.1)
#g = sns.PairGrid(features)
#g = g.map_upper(plt.scatter, edgecolor="w")
#g = g.map_lower(sns.kdeplot)
#g = g.map_diag(plt.hist, edgecolor="b")
#plt.show()

#example of a driver with many events of one type
features.loc[features['F_Harsh turn'].idxmax(), ]
print('features loc:', features.loc)

"""Outliers handling"""
#perform box-cox transformation
import scipy.stats as st

def transform_to_normal(x, min_max_transform = False):
    xt = np.zeros(len(x))
    if np.count_nonzero(x) == 0:
        print("only zero valued values found")
        return x

    valueGreaterThanZero = np.where(x<=0,0,1)
    positives = x[valueGreaterThanZero == 1]
    if(len(positives)> 0):
        xt[valueGreaterThanZero == 1],_ = st.boxcox(positives+1)
    if min_max_transform:
        xt = (xt - np.min(xt)) / (np.max(xt)-np.min(xt))
    return xt

transFeatures = features.apply(lambda x: (transform_to_normal(x, min_max_transform= True)))
print(transFeatures.head())

#scatter plot matrix of transformed features
#g = sns.PairGrid(transFeatures)
#g = g.map_upper(plt.scatter, edgecolor="w")
#g = g.map_lower(sns.kdeplot)
#g = g.map_diag(plt.hist, edgecolor="w")
#plt.show()

#remove outliers
sns.set(style="ticks")

def replace_outliers_with_limit(x, stdFactor = 2.5, normalize=False):
    print(x.name)
    x = x.values
    xt = np.zeros(len(x))
    if np.count_nonzero(x) == 0:
        print("only zero valued values found")
        return x

    xt = transform_to_normal(x)

    xMean, xStd = np.mean(xt), np.std(xt)
    outliers = np.where(xt > xMean + stdFactor*xStd)[0]
    inliers = np.where(xt<= xMean + stdFactor*xStd)[0]
    if len(outliers) > 0:
        print("found outlier with factor:"+str(stdFactor)+ ":"+str(outliers))
        xinline = x[inliers]
        maxInRange = np.max(xinline)
        print("replacing outliers {} with max={}".format(outliers, maxInRange))
        vals = x.copy()
        vals[outliers] = maxInRange
        x = pd.Series(vals)
    else:
        print("No outliers found")
    if normalize:
        #normalize to [0,1]
        x = (x-np.min(x)) / (np.max(x)-np.min(x))
    return x

cleanFeatures = features.apply(lambda x: (replace_outliers_with_limit(x)))
cleanFeatures.head(6)

#g = sns.PairGrid(cleanFeatures)
#g = g.map_upper(plt.scatter, edgecolor="w")
#g = g.map_lower(sns.kdeplot)
#g = g.map_diag(plt.hist, edgecolor="w")
#g = g.set(ylim=((cleanFeatures.min().min()), (cleanFeatures.max().max())))
#plt.show()

#correlations between events
corr = cleanFeatures.corr()

#generate mask for upper triangle
#mask = np.zeros_like(corr, dtype=bool)
#mask[np.triu_indices_from(mask)] = True
fig = plt.figure()
f, ax = plt.subplots(figsize=(7, 7))
cmap = sns.diverging_palette(220, 10, as_cmap=True)

#draw heatmap
sns.heatmap(corr, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink":.5})
plt.show()

"""Modeling"""
#normalize features
minPerFeature = cleanFeatures.min()
maxPerFeature = cleanFeatures.max()

print("min and max values per column before normalization")
for col in range(0, len(cleanFeatures.columns)):
    print("{} range:[{},{}]".format(cleanFeatures.columns[col],minPerFeature[col],maxPerFeature[col]))

normalizedFeatures = (cleanFeatures-cleanFeatures.min())/(cleanFeatures.max()-cleanFeatures.min())
normalizedFeatures.head()
print(normalizedFeatures.head())

#standardize features after box-col
transFeaturesScaled = (transFeatures - transFeatures.mean())/transFeatures.std()
print("")
print("Mean and std before standardization")
for col in range(0, len(transFeatures.columns)):
    print("{} range:[{},{}]".format(transFeatures.columns[col], transFeatures.mean()[col],transFeatures.std()[col]))

normalizedFeatures.head()
print(normalizedFeatures.head())

"""Anomoly detection"""
from sklearn.neighbors import LocalOutlierFactor

X = transFeaturesScaled.values
clf = LocalOutlierFactor(n_neighbors=5)
isOutlier = clf.fit_predict(X)

plt.title("Local Outlier Factor (LOF)", fontsize=20)
a = plt.scatter(X[isOutlier==1, 0], X[isOutlier==1, 1], c='white',
                edgecolors='k', s=40)
b = plt.scatter(X[isOutlier==-1, 0], X[isOutlier==-1, 1], c='red',
                edgecolors='k', s=40)

fig = plt.gcf()
fig.set_size_inches(15,10)
plt.xlabel(normalizedFeatures.columns[0], fontsize=15)
plt.ylabel(normalizedFeatures.columns[1], fontsize=15)
plt.legend([a, b],
           ["normal observations",
           "abnormal observations"],
           loc="upper right", prop={'size':15}, frameon = True)
plt.show()

"""Dimensionality reduction"""
#PCA
from sklearn.decomposition import PCA

pca = PCA(n_components=3)

principalComponents = pca.fit_transform(normalizedFeatures)
column_names = ['principal component {}'.format(i) for i in range(normalizedFeatures.shape[1])]

#plt.bar(x = column_names, height=pca.explained_variance_ratio_)
#plt.title("Percentage of explained variance")
#plt.show()

print("Principal components explained variance ratio: {}.".format(pca.explained_variance_ratio_))

principalDf = pd.DataFrame(data = principalComponents, columns = column_names)

data1 = normalizedFeatures
fig = plt.gcf()
fig.set_size_inches(15,10)

#show correlation matrix of original features and first principal component
#pcAndOriginal = pd.concat([principalDf.iloc[:,0].reset_index(drop=True), normalizedFeatures], axis=1)
#sns.set(style="ticks")
#histplot = pcAndOriginal['principal component 0'].hist(figsize=(2.5, 2.5))
#histplot.set_title("principal component 0 histogram")
#sns.pairplot(pcAndOriginal, y_vars=['principal component 0'], x_vars=['F_Harsh acceleration', 'F_Harsh braking', 'F_Harsh turn'])
#plt.show()

#extract statistics from the fitted distributions
import scipy.stats as st

def fit_distribution_params(series):
    print("Extracting distribution parameters for feature:" + series.name + " (" + str(len(series)) + 'values)')
    xPositive = series[series > 0]

    probs = np.zeros(len(series))
    if(len(xPositive)>0):
        params = st.expon.fit(xPositive)
        arg = params[:-2]
        loc = params[-2]
        scale = params[-1]
        return arg, loc, scale

def calculate_score_for_series(x, fittedParams, verbose = False):
    print("Calculating scores for feature:" + x.name)
    xPositive = x[x>0]

    probs = np.zeros(len(x))
    if(len(xPositive)>0):
        arg = fittedParams[x.name]['arg']
        loc = fittedParams[x.name]['loc']
        scale = fittedParams[x.name]['scale']
        probs[x>0] = st.expon.cdf(xPositive, loc=loc, scale=scale, *arg)
        if verbose:
            probs_df = pd.DataFrame({'Event value':x.values.tolist(), 'Event probability':probs}, index=None)
            probs_df = probs_df.sort_values(by='Event value')
            print(probs_df)
        return probs

#store fitted distribution parameters for later use
fittedParams = {}
for col in normalizedFeatures.columns:
    arg,loc,scale=fit_distribution_params(normalizedFeatures[col])
    fittedParams[col] = {}
    fittedParams[col]['arg'] = arg
    fittedParams[col]['loc'] = loc
    fittedParams[col]['scale'] = scale

import json

print('Fitted parameters:')
print(json.dumps(fittedParams, indent=2))

#commulative distribution/density function
perFeatureScores = normalizedFeatures.apply(calculate_score_for_series, args=(fittedParams, False), axis=0).add_suffix("_CDF")

perFeatureScores.head()
print(perFeatureScores.head())

#plot the fitted distribution

DIST = st.expon

def create_pdf(dist, params, size=10000):
    #seperate parts of parameters
    arg = params[:-2]
    loc = params[-2]
    scale = params[-1]

    start = dist.ppf(0.01, *arg, loc=loc, scale=scale) if arg else dist.ppf(0.01, loc=loc, scale=scale)
    end = dist.ppf(0.99999, *arg, loc=loc, scale=scale) if arg else dist.ppf(0.99999, loc=loc, scale=scale)
    x = np.linspace(start, end, size)
    y = dist.pdf(x, loc=loc, scale=scale, *arg)
    pdf = pd.Series(y, x)

    return pdf

fig, axs = plt.subplots(1, 3, figsize=(15, 6), facecolor='w', edgecolor='k')
fig.subplots_adjust(hspace=.5, wspace=.001)

axs = axs.ravel()
i=0
for col in normalizedFeatures:
    print(col)
    feature = normalizedFeatures[col]
    x = feature.values

    params = DIST.fit(x)

    #seperate parts of parameters
    arg = params[:-2]
    loc = params[-2]
    scale = params[-1]

    pdfForPlot = create_pdf(DIST, params)
    pdfForPlot.plot(lw=2, label='PDF', ax=axs[i])

    feature[feature.values > 0].plot(kind='hist', bins=30, density=True, alpha=0.5, label=feature.name, legend=True, ax=axs[i])
    axs[i].set_ylabel('')
    axs[i].set_xlabel('')

    #Calculate SSE
    yhist, xhist = np.histogram(x, bins=60, density=True)
    xhist = (xhist + np.roll(xhist, -1))[:-1] / 2.0
    histPdf = DIST.pdf(xhist, loc=loc, scale=scale, *arg)
    sse = np.sum(np.power(yhist - histPdf, 2.0))
    print("sse:", sse)
    i+=1

axs[1].set_xlabel('Events per km')
axs[0].set_ylabel('Number of drivers')

plt.show()

"""Create single metric"""

def calculate_joint_score(perFeatureScores):
    driverScores = perFeatureScores
    featureCols = [col for col in driverScores if col.startswith('F_Harsh')]
    driverScores['metric'] = (driverScores[featureCols].sum(axis=1)/3.0)
    driverScores = driverScores.sort_values('metric')
    driverScores['rank'] = (driverScores.metric.rank(method="min")-1/((len(perFeatureScores)-1)*1.0))
    return driverScores

driverScores = calculate_joint_score(perFeatureScores)

from sklearn.linear_model import LinearRegression

#data preparation
X = driverScores['metric'].values.reshape(-1, 1)
y = driverScores['rank'].values

#splitting our data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

#training the model
regressor = LinearRegression()
regressor.fit(X_train, y_train)

#obtain coefficient of determination by calling model with the score() function.
r_sq = regressor.score(X_train, y_train)
print('coefficient of determination:', r_sq)

#print the intercept
print('intercept:', regressor.intercept_)

#print the slope
print('slope:', regressor.coef_)

#predict response and print it
y_pred = regressor.predict(X_test)
print('predicted response:', y_pred, sep='\n')
y_pred2 = regressor.predict(X_train)
print('predicted response_2:', y_pred2, sep='\n')

#plotting regression line
plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, y_pred2, color='b')
plt.xlabel("metric")
plt.ylabel("rank")
plt.title("Linear regression")
plt.show()

#plot for testing dataset
plt.scatter(X_test, y_test, color='red')
plt.plot(X_train, y_pred2, color='blue')
plt.title("Regression testing")
plt.show()

def get_linear_model(perFeatureScores):
    driverScores = calculate_joint_score(perFeatureScores)
    reg = LinearRegression()
    x = driverScores['metric'].values
    y = driverScores['rank'].values
    model = LinearRegression().fit(x, y)
    return model

    print("model:", model)

def get_rank(regModel, metric):
    rank = regModel.predict(metric)
    rank = np.where(rank > 1,1,rank).item()
    rank = np.where(rank < 0,0,rank).item()
    return rank

    print("rank:", rank)

#regModel = get_linear_model(perFeatureScores)
#sampleRank = get_rank(regModel, [[0.5]])
#print("sampleRank:", sampleRank)

#top safest drivers
print(driverScores.head())

#top most risky drivers
print(driverScores.tail())


"""Distributions of CDFs, rank and metric values"""
#sns.set(style="ticks")
#g = sns.PairGrid(driverScores)
#g = g.map_upper(plt.scatter, edgecolor="w")
#g = g.map_diag(plt.hist, edgecolor="w")
#plt.show()







```
